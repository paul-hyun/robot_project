### 2.1 if / elif / else — 장애물 감지 시나리오

```
distance < 0.2  →  비상 정지
0.2 ≤ distance < 1.0  →  감속 운행
distance ≥ 1.0  →  정상 주행
```

````python
distance = 0.05               # 초음파 센서 거리 (m)
is_emergency_stop_pressed = False
````

````python
if distance < 0.2 or is_emergency_stop_pressed:
    print("비상 정지! 장애물이 너무 가깝거나 버튼이 눌렸습니다.")
elif distance < 1.0 and distance >= 0.2:
    print("감속 운행: 장애물 감지됨.")
else:
    print("정상 주행: 경로가 확보되었습니다.")
````

````python
# 다양한 거리 값으로 테스트
test_distances = [0.1, 0.5, 1.5]
print("\n--- 거리별 반응 ---")
for d in test_distances:
    if d < 0.2:
        status = "비상 정지"
    elif d < 1.0:
        status = "감속 운행"
    else:
        status = "정상 주행"
    print(f"  거리 {d}m → {status}")
````

### 2.2 for 루프 — 센서 데이터 배열 처리

| 패턴 | 문법 | 용도 |
|------|------|------|
| 기본 순회 | `for x in list` | 배열 전체 처리 |
| 인덱스 포함 | `for i, x in enumerate(list)` | 방향 + 값 동시 처리 |
| 횟수 반복 | `for _ in range(n)` | 신호 반복 송신 |
| 병렬 순회 | `for a, b in zip(A, B)` | 방향명 + 거리값 매핑 |

````python
lidar_readings = [1.2, 0.8, 2.5, 0.4, 3.0]  # Lidar 5개 방향 거리 (m)
````

````python
# 기본 for 루프
print("--- 기본 순회 ---")
for reading in lidar_readings:
    print(f"  거리: {reading}m")
````

````python
# enumerate: 인덱스와 값을 동시에 처리
print("\n--- enumerate (인덱스 + 값) ---")
for i, reading in enumerate(lidar_readings):
    if i % 2 == 1:
        continue
    if reading < 0.5:
        print(f"  경고: {i}번 방향({reading}m)에 장애물 발견!")
    else:
        print(f"  {i}번 방향: {reading}m 안전")
````

````python
# range: 특정 횟수 반복 / 인덱스 순회
print("\n--- range 활용 ---")
for i in range(3):
    print("  회전 신호 송신 중...")
    print(i)
````

````python
# range: 특정 횟수 반복 / 인덱스 순회
print("\n--- range 활용 ---")
for i in range(0, 3, 1):
    print("  회전 신호 송신 중...")
    print(i)
````

````python
# 0도부터 350도까지 10도 간격으로 스캔
print("\n--- Lidar 각도 스캔 (일부 출력) ---")
for angle in range(0, 360, 10):
    print(angle)
    if angle % 90 == 0:  # 90도 간격으로만 출력
        print(f"  각도 {angle}도 스캔 중...")
````

### 2.3 while 루프 — 로봇 메인 운영 루프

- `continue`: 현재 반복을 건너뛰고 다음 반복으로 진행
- `break`: 루프 전체를 즉시 탈출

> 실제 로봇 소프트웨어에서 `while True:` 루프는 메인 제어 루프의 기본 패턴입니다.

````python
battery_level = 100

# 배터리가 20% 이상일 때만 작동하는 제어 루프
while battery_level > 20:
    print(f"로봇 가동 중... 배터리 잔량: {battery_level}%")
    battery_level -= 15  # 실습에서는 빠르게 소모

    if battery_level == 55:
        print("  중간 점검 중, 다음 사이클로 넘어갑니다.")
        continue  # 아래 코드를 건너뛰고 다음 반복

    if battery_level <= 20:
        print("  배터리 부족으로 기지로 복귀합니다.")
        break  # 루프 전체 탈출

print(f"최종 배터리: {battery_level}%")
````

재사용 가능한 함수 정의와 안전한 예외 처리를 배웁니다.

### 3.1 함수 정의 — 기본 문법

```python
def 함수명(필수인자, 기본값인자=기본값):
    """독스트링 (함수 설명)"""
    return 결과값
```

> **기본값 인자**는 호출 시 생략 가능 → 센서 연결 함수에서 baudrate, timeout 등에 활용

````python
# 기본 함수 정의
def calculate_distance(speed, time):
    """이동 거리 = 속도 × 시간"""
    return speed * time
````

````python
# 기본값이 있는 매개변수
def connect_sensor(port, baudrate=115200, timeout=1.0):
    """시리얼 센서 연결"""
    print(f"  포트 {port} 연결 중 (Baudrate: {baudrate}, Timeout: {timeout}s)")
    return True
````

````python
# 함수 호출
dist = calculate_distance(0.5, 10)
print(f"이동 거리: {dist}m  (speed=0.5m/s, time=10s)")

connect_sensor("/dev/ttyUSB0")                     # 기본값 사용
connect_sensor("/dev/ttyUSB0", baudrate=9600)      # baudrate만 변경
connect_sensor("/dev/ttyUSB0", timeout=2.0)        # timeout만 변경
````

### 3.2 *args와 **kwargs

| 문법 | 수신 형태 | 용도 |
|------|-----------|------|
| `*args` | 튜플 | 개수가 정해지지 않은 위치 인자 |
| `**kwargs` | 딕셔너리 | 개수가 정해지지 않은 키워드 인자 |

````python
# *args: 가변 위치 인자 → 튜플로 수신
def average_readings(*args):
    """여러 센서 값의 평균 계산"""
    if not args:
        return 0.0
    return sum(args) / len(args)

avg = average_readings(1.2, 0.8, 2.5, 0.4, 3.0, 5.0, 7.0, 8.0)
print(f"*args 평균: {avg:.2f}m")
````

````python
# **kwargs: 가변 키워드 인자 → 딕셔너리로 수신
def configure_sensor(**kwargs):
    """유연한 센서 설정값 전달"""
    print(f"  센서 설정: {kwargs}")

configure_sensor(resolution=0.01, port="/dev/ttyUSB0", mode="continuous")
configure_sensor(frequency=10, gain=1.5)
````

````python
# *args와 **kwargs 혼합 사용
def read_sensor_data(sensor_id, timeout=1.0, *args, **kwargs):
    """센서 데이터 읽기"""
    print(f"  센서 ID: {sensor_id}, timeout: {timeout}s")
    print(f"  추가 옵션(args): {args}")
    print(f"  설정값(kwargs): {kwargs}")

read_sensor_data("lidar_01", 2.0, "raw", "debug", frequency=10, mode="scan")
````

### 3.3 람다(Lambda) 함수

- 이름 없는 간단한 함수를 한 줄로 표현
- `filter()`, `map()`, `sort(key=...)` 등과 조합해서 자주 사용

```python
lambda 인자: 표현식
```

````python
class BatteryLowError(Exception):
    """배터리 부족 예외"""
    def __init__(self, level):
        self.level = level
        super().__init__(f"배터리 부족: {level}% (최소 20% 필요)")
    def __str__(self):
        return "안녕하세요."
````

````python
class MotorOverheatError(Exception):
    """모터 과열 예외"""
    pass
````

````python
# 커스텀 예외 활용
def read_sensor_safe(sensor_id, timeout=1.0, **kwargs):
    """센서 데이터 읽기 (예외 처리 포함)"""
    print(f"  센서 {sensor_id} 읽기 시작 (설정: {kwargs})")

    if sensor_id == "thermal":
        raise ValueError("잘못된 센서 데이터 형식 (Checksum Error)")
    elif sensor_id == "motor_port":
        raise HardwareCommError("모터 컨트롤러 연결 끊김")

    return 25.5  # 정상 데이터 반환
````

````python
def main_control_loop(sensor_name):
    """로봇 메인 제어 루프"""
    try:
        data = read_sensor_safe(sensor_name, frequency=10, mode="raw")

    except ValueError as e:
        print(f"  [경고] {e}: 현재 패킷을 버리고 다음 데이터를 기다립니다.")

    except HardwareCommError as e:
        print(f"  [위험] {e}: 비상 정지 시퀀스를 가동합니다.")

    except Exception as e:
        print(f"  [오류] 알 수 없는 오류: {e}")

    else:
        print(f"  [성공] 데이터 정상 수신: {data}")

    finally:
        print("  [finally] 하드웨어 자원 정리 및 안전 모드 확인")
````

````python
print("--- thermal 센서 (ValueError 발생) ---")
main_control_loop("thermal")

print("\n--- motor_port 센서 (HardwareCommError 발생) ---")
main_control_loop("motor_port")

print("\n--- lidar 센서 (정상) ---")
main_control_loop("lidar")
````

## Part 1: 실습 설계 및 요구사항

### 계산기 CLI — 왜 만드나?

로봇 SW의 핵심 구조는 **인지 → 판단 → 동작(Sense-Think-Act)** 의 반복입니다.
계산기 프로그램은 이 **제어 루프(Control Loop)** 의 가장 단순한 형태입니다.

- 사용자 입력을 받고 → **Sense**
- 연산자에 따라 판단하며 → **Think**
- 결과를 출력 → **Act**

오늘 배운 기초 문법들이 실제 로봇의 '지능'을 구현하는 핵심 도구이며,
**복잡한 프로그램을 한 번에 만들지 않고 기능을 하나씩 추가**하는 점진적 구현 방식을 채택합니다.

### 프로그램 요구사항

**계산기 CLI 최종 목표**

| 항목 | 내용 |
|------|------|
| **입력** | 사용자로부터 두 개의 숫자와 한 개의 연산자(`+`, `-`, `*`, `/`)를 입력받음 |
| **연산** | 선택한 연산자에 따라 사칙연산 수행 |
| **반복** | 사용자가 종료를 원할 때까지 계산 과정을 무한히 반복 (로봇의 메인 루프와 동일한 구조) |
| **예외 처리** | 숫자가 아닌 값 입력, 0으로 나누기 등 비정상 상황을 안전하게 처리 |
| **구조화** | 각 연산 로직을 함수로 분리하여 코드의 재사용성 향상 |

### 단계별 구현 순서 — 점진적 개발

복잡한 기능을 한 번에 구현하지 않습니다. 기능 하나씩 완성하며 쌓아가는 방식입니다.

| 단계 | 주제 | 학습 개념 |
|------|------|-----------|
| 1단계 | 기본 입력과 출력 | `input()`, `float()`, `print()` — 자료형 |
| 2단계 | 연산자 선택 기능 추가 | `if-elif-else` — 제어문(조건문) |
| 3단계 | 무한 루프 구현 | `while True`, `break` — 제어문(반복문) |
| 4단계 | 함수를 이용한 모듈화 | `def`, `return` — 함수 |
| 5단계 | 비정상 입력 처리 (최종 완성본) | `try-except` — 예외 처리 |

## Part 2: 단계별 구현

### 1단계 — 기본 입력과 출력

사용자로부터 두 숫자를 받아 더한 결과를 출력합니다.

- `input()`은 모든 입력을 **문자열(str)** 로 반환합니다.
- 산술 연산을 위해 **실수형(float)** 으로 형변환이 필수입니다.
- `float(input(...))` — 입력과 형변환을 한 줄에 처리
- f-string `f"..."` — 변수를 문자열 안에 삽입

```python
# 대화형 셀: 실행 후 입력창에 값 입력
# 1단계: 두 숫자를 입력받아 더하기
num1 = float(input("숫자 1:"))
num2 = float(input("숫자 2:"))

num3 = num1 + num2
print(f"{num1} + {num2} = {num3}")
```

### 2단계 — 연산자 선택 기능 추가

`if-elif-else` 구조로 사용자가 원하는 연산을 선택합니다.

- `if-elif-else`의 마지막 `else`는 **예외 상황을 반드시 처리**해야 합니다.
- 로봇 안전 로직에서 "만약 거리 < 0.5m이면 정지" 구조와 동일합니다.

```python
# 대화형 셀: 실행 후 입력창에 값 입력
# 2단계: 조건문을 활용한 사칙연산 선택
num1 = float(input("숫자 1:"))
num2 = float(input("숫자 2:"))
op = input("연산자 +-*/:")

if op == '+':
    print(f"{num1 + num2}")
elif op == '-':
    print(f"{num1 - num2}")
elif op == '*':
    print(f"{num1 * num2}")
elif op == '/':
    print(f"{num1 / num2}")
else:
    print("잘못된 연산")
```

### 3단계 — 무한 루프 구현

로봇이 전원이 꺼질 때까지 작동하듯, 사용자가 끝낼 때까지 반복합니다.

- `while True` — 로봇 **메인 운영 루프**의 기본 형태
- `break` — 종료 조건을 만족할 때 루프 탈출

```python
# 대화형 셀: 실행 후 입력창에 값 입력
# 3단계: while 루프를 이용한 반복 실행
while True:
    op = input("연산자 +-*/q:")

    if op == 'q':
        break

    num1 = float(input("숫자 1:"))
    num2 = float(input("숫자 2:"))

    if op == '+':
        print(f"{num1 + num2}")
    elif op == '-':
        print(f"{num1 - num2}")
    elif op == '*':
        print(f"{num1 * num2}")
    elif op == '/':
        print(f"{num1 / num2}")
    else:
        print("잘못된 연산")
```

### 4단계 — 함수를 이용한 모듈화

코드의 가독성과 유지보수를 위해 연산 로직을 함수로 분리합니다.

- 함수로 분리 → 나중에 필요할 때마다 **재사용** 가능
- `divide()` 안의 `if b == 0` — 함수 내부에서 선제적 예외 처리

```python
# 대화형 셀: 실행 후 입력창에 값 입력
# 4단계: 함수(Function) 정의 및 호출
def add(a, b):
    return a + b

def sub(a, b):
    return a - b

def mul(a, b):
    return a * b

def div(a, b):
    if b == 0:
        return "입력오류: 0으로 나눌 수 없음."
    return a / b

while True:
    op = input("연산자 +-*/q:")

    if op == 'q':
        break

    num1 = float(input("숫자 1:"))
    num2 = float(input("숫자 2:"))

    if op == '+':
        print(f"{add(num1, num2)}")
    elif op == '-':
        print(f"{sub(num1, num2)}")
    elif op == '*':
        print(f"{mul(num1, num2)}")
    elif op == '/':
        print(f"{div(num1, num2)}")
    else:
        print("잘못된 연산")
```

```python
fnc = {'+': add, '-': sub, '*': mul, '/': div}

while True:
    op = input("연산자 +-*/q:")

    if op == 'q':
        break

    num1 = float(input("숫자 1:"))
    num2 = float(input("숫자 2:"))

    if op in fnc:
        print(fnc[op](num1, num2))
    else:
        print("잘못된 연산")
```

### 5단계 — 비정상 입력 처리 (최종 완성본)

사용자가 숫자가 아닌 문자를 입력해도 프로그램이 강제 종료되지 않도록 합니다.

- `continue` — 루프의 다음 회차로 즉시 넘어감
- `raise` — 조건에 맞을 때 직접 예외를 발생시킴
- `if __name__ == "__main__"` — 모듈로 import될 때와 직접 실행될 때를 구분

```python
# 5단계: try-except를 이용한 견고한 코드 만들기
def calculator():
    """로봇 SW 개발 입문용 계산기 메인 함수"""
    while True:
        try:
            op = input("연산자 +-*/q:")

            if op == 'q':
                break

            if op not in ['+', '-', '*', '/']:
                continue

            num1 = float(input("숫자 1:"))
            num2 = float(input("숫자 2:"))

            if op == '+':
                print(f"{add(num1, num2)}")
            elif op == '-':
                print(f"{sub(num1, num2)}")
            elif op == '*':
                print(f"{mul(num1, num2)}")
            elif op == '/':
                print(f"{div(num1, num2)}")

        except ValueError as e:
            print(e)
        except ZeroDivisionError as e:
            print(e)
        except Exception as e:
            print(e)
    

if __name__ == "__main__":
    calculator()
```

### 1.1 절차형 프로그래밍의 한계

센서가 3개일 때는 변수 3개로 관리할 수 있지만, 10개가 되면?

```python
# 절차형 방식: 센서 3개 → 변수가 무한 증가
sensor1_name = "Front_Ultrasonic"
sensor1_dist = 0.0
sensor2_name = "Left_Ultrasonic"
sensor2_dist = 0.0
sensor3_name = "Right_Ultrasonic"
sensor3_dist = 0.0

# 이 방식이라면 센서가 10개가 되면 변수가 20개 필요!
print(f"센서1: {sensor1_name}, 거리: {sensor1_dist} cm")
print(f"센서2: {sensor2_name}, 거리: {sensor2_dist} cm")
print(f"센서3: {sensor3_name}, 거리: {sensor3_dist} cm")
```

### 1.2 클래스(Class)와 인스턴스(Instance)

- **클래스(Class)**: 설계도 / 틀 — 메모리에 실체가 없는 상태
- **인스턴스(Instance)**: 설계도를 바탕으로 메모리에 실제로 생성된 객체

> 비유: 클래스 = 붕어빵 틀, 인스턴스 = 붕어빵

### 1.3 `__init__`과 `self`

- **`__init__`**: 객체가 처음 만들어질 때 자동으로 실행되는 생성자
- **`self`**: "나 자신"을 가리키는 포인터 — 어느 인스턴스의 변수인지 구별

```python
import random  # 센서 값 시뮬레이션

class DistanceSensor:
    UNIT = "cm"  # 클래스 변수: 모든 센서가 공유

    def __init__(self, name, min_range=0.0, max_range=200.0):
        self.name       = name        # 인스턴스 변수: 센서 이름
        self.min_range  = min_range   # 인스턴스 변수: 최소 범위
        self.max_range  = max_range   # 인스턴스 변수: 최대 범위
        self.last_value = 0.0         # 인스턴스 변수: 최근 측정값

    def get_reading(self):
        # 실제 환경에서는 하드웨어 통신 코드가 들어감
        raw_value = random.uniform(self.min_range, self.max_range)
        self.last_value = round(raw_value, 2)
        print(f"[{self.name}] 측정 완료: {self.last_value} {self.UNIT}")
        return self.last_value
```

```python
# 인스턴스 생성 (두 개의 센서 하드웨어 모델링)
front_sensor = DistanceSensor("Front_Lidar",      10.0, 500.0)
back_sensor  = DistanceSensor("Back_Ultrasonic",   2.0, 150.0)

# 각 센서 인스턴스의 동작 호출 (self는 자동 전달)
val1 = front_sensor.get_reading()
val2 = back_sensor.get_reading()

# 인스턴스 변수 확인
print(f"전방 센서 최근 값: {front_sensor.last_value}")
print(f"후방 센서 최근 값: {back_sensor.last_value}")
```

```python
# 클래스 변수 vs 인스턴스 변수 비교
print(f"클래스 변수 (공유): DistanceSensor.UNIT = {DistanceSensor.UNIT}")
print(f"front_sensor.UNIT  = {front_sensor.UNIT}")   # 클래스 변수 상속
print(f"back_sensor.UNIT   = {back_sensor.UNIT}")    # 클래스 변수 상속
print()
print(f"인스턴스 변수 (각자 독립): front_sensor.name = {front_sensor.name}")
print(f"인스턴스 변수 (각자 독립): back_sensor.name  = {back_sensor.name}")
```

### 1.4 클래스 변수 주의사항

**⚠️ 가변 객체(list, dict)를 클래스 변수로 사용하면** 모든 인스턴스가 공유 →
의도치 않은 데이터 오염 발생!

```python
class BadSensorList:
    readings = []  # ⚠️ 위험: 클래스 변수로 list 사용

class GoodSensorList:
    def __init__(self):
        self.readings = []  # ✅ 안전: 인스턴스 변수로 list 사용

# 위험한 경우 시연
s1 = BadSensorList()
s2 = BadSensorList()
s1.readings.append(10.5)
print(f"s1.readings: {s1.readings}")
print(f"s2.readings: {s2.readings}")  # s2도 영향받음!

# 안전한 경우 시연
g1 = GoodSensorList()
g2 = GoodSensorList()
g1.readings.append(10.5)
print(f"g1.readings: {g1.readings}")
print(f"g2.readings: {g2.readings}")  # g2는 영향받지 않음
```

## Part 2: 상속과 추상화

로봇 센서 클래스 계층 구조를 설계하며 상속(Inheritance), 추상화(Abstraction),
다형성(Polymorphism)을 실습합니다.

### 2.1 추상 베이스 클래스(ABC)와 상속

- **추상 클래스**: 직접 인스턴스를 만들 수 없는 "설계 도면"
- **`@abstractmethod`**: 자식 클래스가 반드시 구현해야 할 메서드를 강제 규정
- **`super()`**: 자식 클래스에서 부모 클래스의 초기화 로직을 실행

```python
from abc import ABC, abstractmethod

# 1단계: 추상 베이스 클래스 정의 (설계도)
class BaseSensor(ABC):
    def __init__(self, name):
        self.name = name
        print(f"[{self.name}] 센서가 초기화되었습니다.")

    @abstractmethod
    def get_data(self):
        """자식 클래스에서 반드시 구현해야 함"""
        pass

# 2단계: LiDAR 클래스 (상속 및 오버라이딩)
class LidarSensor(BaseSensor):
    def __init__(self, name, scan_range):
        super().__init__(name)          # 부모 생성자 먼저 실행
        self.scan_range = scan_range    # 자식만의 속성 추가

    def get_data(self):
        return f"{self.name}: {self.scan_range}도 스캔 완료"
```

```python
# 3단계: Camera / IMU 클래스 (각자 get_data 오버라이딩)
class CameraSensor(BaseSensor):
    def get_data(self):
        return f"{self.name}: RGB 이미지 프레임 획득"

class IMUSensor(BaseSensor):
    def get_data(self):
        return f"{self.name}: 가속도 및 자세 데이터 측정"

# 4단계: 다형성(Polymorphism) 활용 — 제어 루프
sensors = [
    LidarSensor("Hokuyo_LiDAR", 360),
    CameraSensor("Realsense_RGB"),
    IMUSensor("Bosch_IMU"),
]

print("=== 센서 데이터 수집 ===")
for sensor in sensors:
    print(sensor.get_data())  # 각 클래스의 오버라이딩된 메서드 실행
```

### 2.2 추상 클래스 직접 인스턴스화 시도

추상 클래스는 직접 객체를 생성할 수 없습니다.
자식 클래스가 `@abstractmethod`를 구현하지 않으면 에러가 발생합니다.

```python
# 추상 클래스 직접 인스턴스화 → TypeError 발생
try:
    base = BaseSensor("TestSensor")
except TypeError as e:
    print(f"에러 발생: {e}")

# get_data를 구현하지 않은 자식 클래스 → TypeError 발생
class IncompleteSensor(BaseSensor):
    pass  # get_data를 구현하지 않음
    def get_data(self):
        return f"{self.name}: 가속도 및 자세 데이터 측정"

try:
    incomplete = IncompleteSensor("Missing")
except TypeError as e:
    print(f"에러 발생: {e}")
```

### 2.3 다중 상속과 MRO

Python은 두 개 이상의 부모 클래스로부터 상속받는 다중 상속을 지원합니다.
`ClassName.__mro__`로 메서드 탐색 순서(MRO)를 확인할 수 있습니다.

```python
class BluetoothMixin:
    def connect(self):
        return "Bluetooth 연결 완료"

class WifiMixin:
    def connect(self):
        return "WiFi 연결 완료"

class WirelessSensor(BluetoothMixin, WifiMixin, BaseSensor):
    def __init__(self, name):
        super().__init__(name)

    def get_data(self):
        return f"{self.name}: 무선 센서 데이터"

# MRO 확인 — 어떤 순서로 메서드를 탐색하는지 확인
print("WirelessSensor MRO:")
for cls in WirelessSensor.__mro__:
    print(f"  {cls}")

ws = WirelessSensor("BLE_Sensor")
print(ws.connect())  # BluetoothMixin.connect()가 먼저 실행됨
```

### 2.4 합성(Composition) — "has-a 관계"

| 관계 | 설명 | 예시 |
|------|------|------|
| is-a (상속) | LiDAR는 센서이다 | `LidarSensor(BaseSensor)` |
| has-a (합성) | 로봇은 모터를 가진다 | `Robot`이 `Motor` 인스턴스를 속성으로 포함 |

```python
class Motor:
    def __init__(self, name, max_rpm):
        self.name    = name
        self.max_rpm = max_rpm
        self.current_rpm = 0

    def set_speed(self, rpm):
        self.current_rpm = min(rpm, self.max_rpm)
        print(f"[{self.name}] 속도 설정: {self.current_rpm} RPM")

class Robot:
    """합성(Composition): Robot has-a Motor, has-a LidarSensor"""
    def __init__(self, name):
        self.name   = name
        self.motor  = Motor("MainMotor", max_rpm=300)       # has-a Motor
        self.sensor = LidarSensor("FrontLidar", scan_range=180)  # has-a Sensor

    def move(self, speed):
        self.motor.set_speed(speed)

    def scan(self):
        return self.sensor.get_data()

robot = Robot("ROS-Bot")
robot.move(150)
print(robot.scan())
```

```python
# TODO: 추상 클래스를 이용한 센서 계층 구조를 완성하세요.
# 요구사항:
#   - TemperatureSensor는 BaseSensorABC를 상속받아야 함
#   - get_data(self)는 "현재 온도: 25.0°C" 형태의 문자열을 반환
#   - UltrasonicSensor.get_data(self)는 "거리: 45.2 cm" 형태 반환
#
# Hint: @abstractmethod 데코레이터, super().__init__() 호출

class BaseSensorABC(ABC):
    def __init__(self, name):
        self.name = name

    @abstractmethod
    def get_data(self):
        pass

class TemperatureSensor(BaseSensorABC):
    def __init__(self, name):
        super().__init__(name)

    def get_data(self):
        return "현재 온도: 25.0°C"

class UltrasonicSensor(BaseSensorABC):
    def __init__(self, name):
        super().__init__(name)

    def get_data(self):
        return "거리: 45.2 cm"

# 완성 후 주석 해제
sensors_list = [TemperatureSensor("온도센서"), UltrasonicSensor("초음파센서")]
for s in sensors_list:
    print(s.get_data())
```

```python
# TODO: 합성(Composition)을 활용한 로봇 클래스를 완성하세요.
# 요구사항:
#   - RobotSystem은 TemperatureSensor와 Motor를 속성으로 가짐 (has-a 관계)
#   - report(self) 메서드는 온도 센서 데이터와 모터 상태를 모두 출력
#
# Hint: __init__ 내에서 TemperatureSensor("내부온도")와 Motor(100) 인스턴스 생성

class RobotSystem:
    def __init__(self, name):
        self.name = name
        self.temp_sensor = TemperatureSensor("내부온도")
        self.motor = Motor(100)

    def report(self):
        print(self.temp_sensor.get_data())
        print(self.motor.status())

# 완성 후 주석 해제
bot = RobotSystem("CleanBot")
bot.report()
```
